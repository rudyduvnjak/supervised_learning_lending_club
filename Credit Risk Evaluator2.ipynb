## Identify Categorical Columns

```python

# Split the columns into quantitative and categorical
cat_cols_train = [col for col in train_df.columns if col not in train_df.describe().columns]
cat_cols_test = [col for col in test_df.columns if col not in test_df.describe().columns]
print(cat_cols_train)
```

## Data Conversion Into Numerical Mode

```python
# Convert categorical data to numeric and separate target feature for training data
train_df2 = pd.get_dummies(train_df, columns=cat_cols_train)
train_df2.head()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
test_df2 = pd.get_dummies(test_df, columns=cat_cols_test)
test_df2.head()
```

```python
# Check for column matching between train_df2 and test_df2
check_column_matching()
```

```python
# Fix the missing values for the debt_settlement_flag columns we get_dummies'd earlier
# by adding missing dummy variables to testing set
test_df2["debt_settlement_flag_Y"] = 0
check_column_matching()
```

## Check for Null values

```python
# add missing dummy variables to testing set
null_variables = pd.isnull(test_df2)
test_df2.fillna(0)
```

```python
# Label target columns
target = "loan_status"

```

## Split the Data for Training and Test

```python
# Split the training data
X_train  = train_df2.drop(columns=[target])
y_train = train_df2[[target]].values.ravel()
#
print(X_train.shape, y_train.shape)
```

```python
# Split the testing data
X_test  = test_df2.drop(columns=[target])
y_test = test_df2[[target]].values.ravel()
#
print(X_test.shape, y_test.shape)
```

```python
X_train.head(2)
```

```python
X_test.head(2)
```

```python
# Identify missing columns in numerical data
missing_cols = set( X_train.columns ) - set( X_test.columns )
print(missing_cols)
# Add missing column to X_test and add value of 0
for col in missing_cols:
    X_test[col] = 0

# Align the order of the columns 
X_train, X_test = X_train.align(X_test, axis=1)

# Verify column alignment
X_train.columns==X_test.columns
```

## Complete Logistic Regression and Random Forest Classifier 

```python
# Train the Logistic Regression model on the unscaled data and print the model score
logisticRegr = LogisticRegression(solver='lbfgs', max_iter=2000,random_state=0)
logisticRegr.fit(X_train, y_train)
print("Logistic Regression Train Score", logisticRegr.score(X_train, y_train))
print("LogisticRegression Test Score: ", logisticRegr.score(X_test, y_test))
```

```python
# Train a Random Forest Classifier model and print the model score
randomForestClass = RandomForestClassifier(random_state=42)
randomForestClass.fit(X_train, y_train)
print("Random Forest Clasifer Train Score", randomForestClass.score(X_train, y_train))
print("Random Forest Classifier Test Score: ", randomForestClass.score(X_test, y_test))
```

## Scale the Data
*I think that Logistic Regression will perform better after scalling data
*Logistic Regression should perform better because data will be scalled, hence more uniform and more refined
*Random Forest Classifier should improve score, but not as much as Logistic Regression due to nature of method

```python
# Scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
LR_scaled = LogisticRegression(solver='sag', max_iter=3000, random_state=0)
LR_scaled.fit(X_train_scaled, y_train)
print("Logistic Regression  Train Scaled Score: ", LR_scaled.score(X_train_scaled, y_train))
print("Logistic Regression Test Scaled Score: ", LR_scaled.score(X_test_scaled, y_test))
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
random_forest_scaled = RandomForestClassifier(random_state=100)
random_forest_scaled.fit(X_train_scaled, y_train)
print("RandomForestClassifier scaled score: ", random_forest_scaled.score(X_train_scaled, y_train))
print("RandomForestClassifier scaled score: ", random_forest_scaled.score(X_test_scaled, y_test))
```

## Scaled Data Evaluation
*Even though Logistic Regression improved, it didn't improve substantially after data scaling
*Test Score improved with logistic Regression
*Random Forest Classifier didn't improve substantially
*It appears that data isn't properly balanced 
*I think that result would be substantially better if all data was combined and then train_test_split
